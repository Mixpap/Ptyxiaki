## Initial Estimation of Optical Depth
 Using the known abundance ratio of $\frac{^{12}CO}{^{13}CO}=ab=60$ and the relationship between the ratio of brightness temperatures of the two isotopes: $\frac{T_{b}(^{12}CO)}{T_{b}(^{13}CO)}=\frac{1}{1-e^\tau ({^{12}CO}/ab)}$ we can estimate the optical depth map of $^{12}CO$ 
$$
\tau [y,x] _{^{12}CO}=-ab \ln (1-\frac{T_{b}[y,x] _{^{13}CO} }{T_{b}[y,x] _{^{12}CO}}      )
$$

```python
def initial_est(map12,map13,abundance):
    T0=np.zeros(map13.shape)
    Y,X=T0.shape[0],T0.shape[1]
    ab = abundance #kanonika 60
    for y in range(Y):
        for x in range(X):
            if (np.isnan(map13[y,x]) or np.isnan(map12[y,x]) or (map12[y,x]==0.0)):
                        T0[y,x]=np.nan
            else:
                T0[y,x]=-ab*np.log(1-map13[y,x]/map12[y,x])
    return T0
```

```python
T0 = initial_est(C18m,CO13m,6.0)
#T0 = initial_est(CO12m,CO13m)
```

## Final Estimation of Optical Depth
Using the previous estimated map of OD we evaluate the Newton-Raphson iterative method to optimize the result.
$$
\tau [y,x] _{n+1} = \tau [y,x]_n-\frac{(\frac{(T_{b}[y,x] _{^{12}CO} }{T_{b}[y,x] _{^{13}CO}})\times (1-e^{-\tau [y,x]_n/ab}))-(1-e^{-\tau [y,x]_n})}{(\frac{T_{b}[y,x] _{^{12}CO} }{T_{b}[y,x] _{^{13}CO}} \times \frac{e^{-\tau [y,x]_n/ab}}{ab})- e^{-\tau [y,x]_n}}
$$

* We loop until a tolerance ratio is achieved

```python
def final_est(map12,map13,T0,abundance):
    Y,X=T0.shape[0],T0.shape[1]
    T=T0.copy()
    loops=[]
    ab=abundance #60 
    tol=0.00001
    maxi=20
    for y in range(Y):
        for x in range(X):
            if (np.isnan(T[y,x]) or (map13[y,x]==0.0)):
                T[y,x]=np.nan
            else:
                i=0
                f=0.1
                while f>tol:                   
                    A=map12[y,x]/map13[y,x]
                    B=1.0-np.exp(-T[y,x]/ab)
                    C=1.0-np.exp(-T[y,x])
                    D=np.exp(-T[y,x]/ab)/ab
                    F=np.exp(-T[y,x])
                    temp=T[y,x]
                    T[y,x]=T[y,x]-(A*B-C)/(A*D-F)
                    f=T[y,x]-temp
                    i=i+1
                    
                    if i>=maxi:
                        break
                loops.append(i)
    return T,loops
```

```python
T,k=final_est(C18m,CO13m,T0,6.0)
```

```python
np.array(k).max()
```

```python
figsize(15,15)
plt.title('Optical Depth Map')
plt.imshow(T,origin='low',cmap='coolwarm',vmax=160)
plt.colorbar()
```

#Optical Thin
$$
\tau [y,x] < 1
$$

```python
def thin_est(T):
    Y,X=T.shape[0],T.shape[1]
    Thin=T.copy()
    for y in range(Y):
        for x in range(X):
            if Thin[y,x]>1:
                Thin[y,x]=np.nan
    return Thin
```

```python
Thin=thin_est(T)
```

```python
figsize(8,8)
plt.title('Optical Thin Map')
plt.imshow(Thin,cmap='coolwarm',origin='low')
```
