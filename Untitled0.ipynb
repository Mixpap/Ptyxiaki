```python
c18=fits.getdata('data/cubes/C18O.fits')
w18 = wcs.WCS('data/cubes/C18O.fits')
Z,Y,X =np.shape(c18)

def spectrum(y,x):
    return c18[:,y,x]

def func(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

def gauss_fit(xx,ym,xm):
    popt, pcov = curve_fit(func, xx, spectrum(ym,xm))
    return func(xx,popt[0],popt[1],popt[2])
```

```python
ym,xm=433,126
xx=np.linspace(0,1,61)
s=spectrum(ym,xm)
s_error=0.4*np.sqrt(np.abs(s))
f1=gauss_fit(xx,ym,xm)
plt.plot(xx,s,'ko')
plt.errorbar(xx,s,s_error)
plt.plot(xx,f1)
#plt.annotate()
```

```python
def model(x, s,s_error): 
    a = pm.Uniform('a', 0.5,15.0)
    x0 = pm.Uniform('x0', 0.3, 1.0)
    sigma = pm.Uniform('sigma', 0.01, 0.75)
    
    @pm.deterministic(plot=False)
    
    def gauss(x=x, a=a,x0=x0,sigma=sigma):
        #s=0.05
        return func(x,a,x0,sigma)
    
    f = pm.Normal('f', mu=gauss, tau=1.0/s_error**2, value=s, observed=True)
    
    return locals()
```

```python
MDL = pm.MCMC(model(xx,s,s_error))
MDL.sample(50000, 4000, 1)
```

```python
#============~PLOTTING RESULTS~==================
figsize(18,10)
#============~Scipy curve_fit Testing~==========
popt, pcov = curve_fit(func, xx, s)
s_curve_fit=func(xx,popt[0],popt[1],popt[2])
plt.plot(xx,s_curve_fit,label='Scipy Curve Fit')

#================~Data~=====================
plt.plot(xx,s,'b', marker='None', ls='--', lw=1, label='True')
plt.errorbar(xx,s,s_error, color='r', marker='.', label='Observed')

#==============~MCMC fit~================
s_min = MDL.stats()['gauss']['quantiles'][2.5]
s_max = MDL.stats()['gauss']['quantiles'][97.5]
s_fit = MDL.stats()['gauss']['mean']

plt.plot(xx,s_fit,'k', ms=5, mew=1, label='MCMC Fit')
plt.fill_between(xx, s_min, s_max, color='0.5', alpha=0.5)

#=============~MidPoints~====================
s_mid=s_fit.max()/2.0
x0=MDL.stats()['x0']['mean']
sigma=MDL.stats()['sigma']['mean']
a=MDL.stats()['a']['mean']
x_mid1=x0+sigma*np.sqrt(2*np.log(a/s_mid))
x_mid2=x0-sigma*np.sqrt(2*np.log(a/s_mid))

plt.axhline(y=s_mid,color='r',ls='dashed')
plt.axvline(x=x_mid1,color='r',ls='dashed')
plt.axvline(x=x_mid2,color='r',ls='dashed')


plt.legend()
```
