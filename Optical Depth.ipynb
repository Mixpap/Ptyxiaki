## Initial Estimation of Optical Depth
 Using the known abundance ratio of $\frac{^{12}CO}{^{13}CO}=77$ and the relationship between the ratio of brightness temperatures of the two isotopes: $\frac{T_{b}(^{12}CO)}{T_{b}(^{13}CO)}=\frac{1}{1-e^\tau ({^{12}CO}/ab)}$ we can estimate the optical depth map of $^{12}CO$ 
$$
\tau [y,x] _{^{12}CO}=-ab \ln (1-\frac{T_{b}[y,x] _{^{13}CO} }{T_{b}[y,x] _{^{12}CO}}      )
$$

```python
tau12 = functions.initial_est(CO12pnr,CO13pnr,77.0)
tau13 = functions.initial_est(CO13pnr,C18pnr,8.0)
```

## Final Estimation of Optical Depth
Using the previous estimated map of OD we evaluate the Newton-Raphson iterative method to optimize the result.
$$
\tau [y,x] _{n+1} = \tau [y,x]_n-\frac{(\frac{(T_{b}[y,x] _{^{12}CO} }{T_{b}[y,x] _{^{13}CO}})\times (1-e^{-\tau [y,x]_n/ab}))-(1-e^{-\tau [y,x]_n})}{(\frac{T_{b}[y,x] _{^{12}CO} }{T_{b}[y,x] _{^{13}CO}} \times \frac{e^{-\tau [y,x]_n/ab}}{ab})- e^{-\tau [y,x]_n}}
$$

* Max 10 loops

```python
tau12_f=functions.final_est(CO12pnr,CO13pnr,tau12,77.0)
tau13_f=functions.final_est(CO13pnr,C18pnr,tau13,8.0)
```

```python
functions.plot_maps(maps=[tau12_f,tau13_f],titles=[r'$\tau ^{12}$',r'$\tau ^{13}$'],coords=w12)
```

```python
def map_show(line):
    zlen=61.
    xx=np.linspace(0,1,zlen)
    T=CO12[line,::]
    ym,xm=np.where(T==T.max())
    s=functions.Spectra(CO12,ym[0],xm[0])
    der1=np.diff(s)
    der2=np.diff(np.diff(s))
    der3=np.diff(np.diff(np.diff(s)))
    m = np.ones(len(s), dtype=bool)
    ind1=np.argsort(der2)[0]+1
    ind2=np.argsort(der2)[1]+1
    if np.abs(ind1-ind2)<2:
        ind2=np.argsort(der2)[2]+1
    m[ind1:ind2]=False
    m[ind2:ind1]=False
    popt, pcov = curve_fit(functions.gaussian, xx[m], s[m])
    #popt, pcov = curve_fit(twogaussians, xx, s)
    FWHM=2.355*np.abs(popt[2])
    
    ax1 = pywcsgrid2.subplot(211, wcs=w18)
    ax1.imshow(T,origin='low',cmap='coolwarm')
    #ax1.set_title('$C^{18}O$')
    ax1.add_compass(loc=4,color='black')
    ax1.axvline(x=xm,color='r',ls='dashed')
    ax1.annotate(xm[0],(xm+10,10),color='r')
    ax1.axhline(y=ym,color='r',ls='dashed')
    ax1.annotate(ym[0],(10,ym+10),color='r')
    ax1.annotate(r'$T_{max}=$%0.2f'%(T[ym,xm]),(xm+30,ym+10),color='r',size='large')
    
    ax2=plt.subplot(212)
    ax2.plot(xx,s,label='True Data')
    ax2.plot(xx[m],s[m],'ko',label='Masked Data')
    ax2.plot(xx,functions.gaussian(xx,popt[0],popt[1],popt[2]),label='Fit to Masked Data')
    ax2.annotate(r'Fit Parameters: $A=$%0.3f, $x_0=$%0.3f, $\sigma=$%0.3f'%(popt[0],popt[1],popt[2]),(0.7,popt[0]/2-3))
    #ax2.plot(xx,twogaussians(xx,popt[0],popt[1],popt[2],popt[3],popt[4],popt[5]))
    ax2.axhline(y=popt[0]/2,color='r',label='HalfMaximum of fit')
    ax2.annotate(r'$FWHM=$%0.3f'%(FWHM),(0.8,popt[0]/2+0.5))
    #ax2.plot(np.linspace(0,1,zlen-1),der1,label='First Derivative')
    ax2.plot(np.linspace(0,1,zlen-2),der2,label='Second Derivative')
    ax2.plot(xx,np.where(m==False,popt[0],0.0),label='Mask',alpha=0.3,ls='dashed')

    ax2.legend()
  
```
