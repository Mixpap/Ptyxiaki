## Initial Estimation of Optical Depth\n",
Using the known abundance ratio of $\frac{^{12}CO}{^{13}CO}=77$ and the relationship between the ratio of brightness temperatures of the two isotopes: $\frac{T_{b}(^{12}CO)}{T_{b}(^{13}CO)}=\frac{1}{1-e^\tau ({^{12}CO}/ab)}$ we can estimate the optical depth map of $^{12}CO$
      $$\tau [y,x] _{^{12}CO}=-ab \ln (1-\frac{T_{b}[y,x] _{^{13}CO} }{T_{b}[y,x] _{^{12}CO}}) $$

```python
tau12 = functions.initial_est(CO12pnr,CO13pnr,functions.ab12CO)
tau13 = functions.initial_est(CO13pnr,C18pnr,functions.ab13CO)
```

## Final Estimation of Optical Depth
Using the previous estimated map of OD we evaluate the Newton-Raphson iterative method to optimize the result.
$$
\tau [y,x] _{n+1} = \tau [y,x]_n-\frac{(\frac{(T_{b}[y,x] _{^{12}CO} }{T_{b}[y,x] _{^{13}CO}})\times (1-e^{-\tau [y,x]_n/ab}))-(1-e^{-\tau [y,x]_n})}{(\frac{T_{b}[y,x] _{^{12}CO} }{T_{b}[y,x] _{^{13}CO}} \times \frac{e^{-\tau [y,x]_n/ab}}{ab})- e^{-\tau [y,x]_n}}
$$

* Max 10 loops

###Υπολογισμός FWHM 
Για να υπολογίσουμε τις παραμέτρους του Gauss-Fitting χρησιμοποιούμε τη συνάρτητη **curve_fit** της βιβλιοθήκης **Scipy** η οποία κάνει non-linear least-square fit της συνάρτηση μας.
Για καλύτερη απόδοση και για να αποφύγουμε λάθος local-minimum ο αλγόριθμος εχει 3 κομμάτια:
1. Δημιουργεί μια μάσκα στο φάσμα του $^{12}CO$ (απο τη 2η παράγωγο) έτσι ώστε να αποφύγει το self-absorbed κομμάτι και βρίσκει το gaussian fit *με αρχική συνθήκη το μέγιστο σημείο (πλάτος και $x_0$).
2. Βρίσκει το fit στο φάσμα του $^{13}CO$ χρησιμοποιώντας τo αποτέλεσμα του βήματος 1 σαν αρχική συνθήκη.
3. Βρίσκει το fit στο φάσμα του $C^{18}O$ χρησιμοποιώντας τo αποτέλεσμα του βήματος 2 σαν αρχική συνθήκη.
> Το έχω τρέξει μέχρι στιγμής για κάθε pixel καθε αυτό, δηλαδή όχι με τη μέση τιμή 3X3.

```python
def f(gf):
    critical_turbulence=4.
    index_deviation=[10,2] #Max and Min Index Deviation for Second Derivative Mask
    Y=np.shape(C18)[1]
    X=np.shape(C18)[2]
    xx=functions.velocity
    Peak12=np.zeros((Y,X))
    Peak13=np.zeros((Y,X))
    Peak18=np.zeros((Y,X))
    PeakSD12=np.zeros((Y,X))
    PeakSD13=np.zeros((Y,X))
    PeakSD18=np.zeros((Y,X))
    FWHM12=np.zeros((Y,X))
    FWHM13=np.zeros((Y,X))
    FWHM18=np.zeros((Y,X))
    FWHMSD12=np.zeros((Y,X))
    FWHMSD13=np.zeros((Y,X))
    FWHMSD18=np.zeros((Y,X))
    Integral12=np.zeros((Y,X))
    Integral12full=np.zeros((Y,X))
    Integral12out=np.zeros((Y,X))

    for y in range(Y):
        for x in range(X):
            if (not np.ma.is_masked(CO12pnr[y,x])): #((tau13_f[y,x]<ct) and 
                s12=functions.Spectra(CO12r,y,x)
                Integral12full[y,x]=s12.sum()
                
                s13=functions.Spectra(CO13r,y,x)
                s18=functions.Spectra(C18r,y,x)
                der2=np.diff(s12,2.)
                #---------------Mask------------------
                m = np.ones(len(s12), dtype=bool)
                ind1=np.argsort(der2)[0]+1
                ind2=np.argsort(der2)[1]+1
                i=0
                while (np.abs(ind1-ind2)>index_deviation[0] or np.abs(ind1-ind2)<index_deviation[1]):
                    #print i
                    ind2=np.argsort(der2)[i]+1
                    i=i+1
                m[ind1:ind2]=False
                m[ind2:ind1]=False
                #-----------End-of-Mask---------------
                
                fit12,fit13,fit18=[False,False,False]
                
                #============Fitting================================
                #-----------CO12-----------------------
                try:
                    popt12, pcov12 = curve_fit(functions.gaussian, xx[m], s12[m],p0=[s12.max(),xx[np.argmax(s12)],1.5],diag=(0.01,0.01))
                except:
                    #popt12,pcov12=np.zeros((1,3)),np.zeros((3,3))
                    continue  #next loop
                sd12= np.sqrt(np.diag(pcov12)) 
                fit12 = (sd12<gf*np.abs(popt12)).all() #Good Fit?
                
                if fit12:
                    Peak12[y,x]=popt12[1]
                    PeakSD12[y,x]=sd12[1]
                    FWHM12[y,x]=2.355*np.abs(popt12[2])
                    FWHMSD12[y,x]=2.355*np.abs(sd12[2])
                else:
                    continue
                #----------CO13--------------------------------
                if (not np.ma.is_masked(CO13pnr[y,x])):
                    try:
                        popt13, pcov13 = curve_fit(functions.gaussian, xx, s13,p0=[0.25*popt12[0],popt12[1],popt12[2]],diag=(0.01,0.01))
                    except:
                        #popt13,pcov13=np.zeros((3)),np.zeros((3,3))
                        continue  #next loop
                    sd13= np.sqrt(np.diag(pcov13))      #Standard Deviation
                    fit13 = (sd13<gf*np.abs(popt13)).all() #Good Fit?

                    if fit13:
                        Peak13[y,x]=popt13[1]
                        PeakSD13[y,x]=sd13[1]
                        FWHM13[y,x]=2.355*np.abs(popt13[2])
                        FWHMSD13[y,x]=2.355*np.abs(sd13[2])
                        if (FWHM12[y,x]>critical_turbulence):
                            Integral12[y,x]=s12[xx<popt13[1]-FWHM13[y,x]/2].sum()+s12[xx>popt13[1]+FWHM13[y,x]/2].sum() 
                            Integral12out[y,x]=s12.sum()
                    else:
                        continue
                #----------C18O-------------------------------------
                if (not np.ma.is_masked(C18pnr[y,x])):
                    try:
                        popt18, pcov18 = curve_fit(functions.gaussian, xx, s18,p0=[0.25*popt13[0],popt13[1],popt13[2]],diag=(0.005,0.005))
                    except:
                        #popt18,pcov18=np.zeros((3)),np.zeros((3,3))
                        continue  #next loop
                    sd18= np.sqrt(np.diag(pcov18))      #Standard Deviation
                    fit18 = (sd18<gf*np.abs(popt18)).all() #Good Fit?

                    if fit18:
                        Peak18[y,x]=popt18[1]
                        PeakSD18[y,x]=sd18[1]
                        FWHM18[y,x]=2.355*np.abs(popt18[2])
                        FWHMSD18[y,x]=2.355*np.abs(sd18[2])
                

    return Peak12,PeakSD12,FWHM12,FWHMSD12,Peak13,PeakSD13,FWHM13,FWHMSD13,Peak18,PeakSD18,FWHM18,FWHMSD18,Integral12full*functions.vres,Integral12out*functions.vres,Integral12*functions.vres
    #return Integral12full,Integral12out,Integral12
```

```python
def Mas(y,x):
    s=functions.Spectra(CO12r,y,x)
    d=2000.*functions.pac
    X=1./(8.5e-5)

    A=1.28e14
    T10=functions.h *115.271*10**9 /functions.k
    T32=functions.h *functions.v_co12_j32*10**9 /functions.k
    Tbg=2.7
    Z=0.36156*(Tx12[y,x]+0.922)
    BB=np.exp(T10/Tx12[y,x])
    B=(1.-np.exp(-T32/Tx12[y,x]))
    C=(1./(np.exp(T32/Tx12[y,x])-1.)-1./(np.exp(T32/Tbg)-1.))**(-1)
    D=tau12_f[y,x]/(1.-np.exp(-tau12_f[y,x]))
    integ=s.sum()*1000.*functions.vres

    N=A*Z*BB*B*C*D*integ
    M=N*2.8*(3.35e-27)*(7.7/206265)**2 *X *d**2

    return M/functions.Msol

def Integral_to_mass(integral,Tx,tau):
    d=2000.*functions.pac
    X=1./(8.5e-5)
    A=1.28e14
    T10=functions.h *115.271*10**9 /functions.k
    T32=functions.h *functions.v_co12_j32*10**9 /functions.k
    Tbg=2.7
    Z=0.36156*(Tx+0.922)
    BB=np.exp(T10/Tx)
    B=(1.-np.exp(-T32/Tx))
    C=(1./(np.exp(T32/Tx)-1.)-1./(np.exp(T32/Tbg)-1.))**(-1)
    D=tau/(1.-np.exp(-tau12))
    integ=integral*1000.

    N=A*Z*BB*B*C*D*integ
    M=N*2.8*(3.35e-27)*(7.7/206265)**2 *X *d**2

    return M/functions.Msol
```

###3X3 Integral Est.

```python
def f9(gf):
    index_deviation=[10,2] #Max and Min Index Deviation for Second Derivative Mask
    Y=np.shape(C18)[1]
    X=np.shape(C18)[2]
    xx=functions.velocity
    Peak12=np.zeros((Y,X))
    Peak13=np.zeros((Y,X))
    Peak18=np.zeros((Y,X))
    PeakSD12=np.zeros((Y,X))
    PeakSD13=np.zeros((Y,X))
    PeakSD18=np.zeros((Y,X))
    FWHM12=np.zeros((Y,X))
    FWHM13=np.zeros((Y,X))
    FWHM18=np.zeros((Y,X))
    FWHMSD12=np.zeros((Y,X))
    FWHMSD13=np.zeros((Y,X))
    FWHMSD18=np.zeros((Y,X))
    Integral12=np.zeros((Y,X))

    for y in range(Y)[1:-1]:
        for x in range(X)[1:-1]:
            if (not np.ma.is_masked(CO12pnr[y,x])): #((tau13_f[y,x]<ct) and 
                s12=functions.Spectra9(CO12,y,x)[0]
                s13=functions.Spectra9(CO13,y,x)[0]
                s18=functions.Spectra9(C18,y,x)[0]
                der2=np.diff(s12,2.)
                #---------------Mask------------------
                m = np.ones(len(s12), dtype=bool)
                ind1=np.argsort(der2)[0]+1
                ind2=np.argsort(der2)[1]+1
                i=0
                while (np.abs(ind1-ind2)>index_deviation[0] or np.abs(ind1-ind2)<index_deviation[1]):
                    #print i
                    ind2=np.argsort(der2)[i]+1
                    i=i+1
                m[ind1:ind2]=False
                m[ind2:ind1]=False
                #-----------End-of-Mask---------------
                
                fit12,fit13,fit18=[False,False,False]
                
                #============Fitting================================
                #-----------CO12-----------------------
                try:
                    popt12, pcov12 = curve_fit(functions.gaussian, xx[m], s12[m],p0=[s12.max(),xx[np.argmax(s12)],1.5],diag=(0.01,0.01))
                except:
                    #popt12,pcov12=np.zeros((1,3)),np.zeros((3,3))
                    continue  #next loop
                sd12= np.sqrt(np.diag(pcov12)) 
                fit12 = (sd12<gf*np.abs(popt12)).all() #Good Fit?
                
                if fit12:
                    Peak12[y,x]=popt12[1]
                    PeakSD12[y,x]=sd12[1]
                    FWHM12[y,x]=2.355*np.abs(popt12[2])
                    FWHMSD12[y,x]=2.355*np.abs(sd12[2])
                else:
                    continue
                #----------CO13--------------------------------
                if (not np.ma.is_masked(CO13pnr[y,x])):
                    try:
                        popt13, pcov13 = curve_fit(functions.gaussian, xx, s13,p0=[0.25*popt12[0],popt12[1],popt12[2]],diag=(0.01,0.01))
                    except:
                        #popt13,pcov13=np.zeros((3)),np.zeros((3,3))
                        continue  #next loop
                    sd13= np.sqrt(np.diag(pcov13))      #Standard Deviation
                    fit13 = (sd13<gf*np.abs(popt13)).all() #Good Fit?

                    if fit13:
                        Peak13[y,x]=popt13[1]
                        PeakSD13[y,x]=sd13[1]
                        FWHM13[y,x]=2.355*np.abs(popt13[2])
                        FWHMSD13[y,x]=2.355*np.abs(sd13[2])
                        Integral12[y,x]=s12[xx<popt13[1]-FWHM13[y,x]/2].sum()+s12[xx>popt13[1]+FWHM13[y,x]/2].sum() 
                    else:
                        continue
                #----------C18O-------------------------------------
                if (not np.ma.is_masked(C18pnr[y,x])):
                    try:
                        popt18, pcov18 = curve_fit(functions.gaussian, xx, s18,p0=[0.25*popt13[0],popt13[1],popt13[2]],diag=(0.1,0.1))
                    except:
                        #popt18,pcov18=np.zeros((3)),np.zeros((3,3))
                        continue  #next loop
                    sd18= np.sqrt(np.diag(pcov18))      #Standard Deviation
                    fit18 = (sd18<gf*np.abs(popt18)).all() #Good Fit?

                    if fit18:
                        Peak18[y,x]=popt18[1]
                        PeakSD18[y,x]=sd18[1]
                        FWHM18[y,x]=2.355*np.abs(popt18[2])
                        FWHMSD18[y,x]=2.355*np.abs(sd18[2])
                

    return Peak12,PeakSD12,FWHM12,FWHMSD12,Peak13,PeakSD13,FWHM13,FWHMSD13,Peak18,PeakSD18,FWHM18,FWHMSD18,Integral12
```
